### import required packages 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

### load the data 

```python
df = pd.read_csv('NewspaperData.csv')
df_delivery = pd.read_csv('delivery_time.csv')
df_wc = pd.read_csv('WC_AT.csv')
```

### pre-processing 

```python
# split the data in x and y
x = df.drop(['Newspaper','sunday'], axis=1)
y = df['sunday'] 
```

```python
x_delivery = df_delivery.drop('Sorting Time',axis=1)
y_delivery = df_delivery['Sorting Time']
```

```python
x_wc = df_wc.drop('AT',axis=1)
y_wc = df_wc['AT']
```

### train a DL model 

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()

model.add(Dense(units=10 , input_shape=[x.shape[1]] ,activation='relu'))

model.add(Dense(units=20,activation='relu') )
model.add(Dense(units=10))

model.add(Dense(units=1))

optimizer_function = tf.optimizers.Adam()
loss_function = tf.losses.MeanSquaredError()

model.compile(optimizer = optimizer_function , loss=loss_function)
```

```python
# build the model
model.fit(x, y, epochs=100)
```

```python
model.predict(np.array([391]))
```

```python
model= Sequential()
model.add(Dense(units=20,input_shape=[x.shape[1]]))
model.add(Dense(units=10,activation='relu'))
model.add(Dense(units=10,activation='relu'))

model.add(Dense(units=1,activation='linear'))

optimizer_function = tf.optimizers.Adam()
loss_function = tf.losses.MeanAbsolutePercentageError()
model.compile( optimizer= optimizer_function ,loss=loss_function)
```

```python
# build the model
model.fit(x_delivery, y_delivery, epochs=100)
```

```python
model.predict(np.array([21]))
```
