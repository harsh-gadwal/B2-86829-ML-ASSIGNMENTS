In Churn.csv dataset you have to predict that customer is exited or not based on various factors given in dataset.
Implement a KNN,SVM and Naive Bayes model and calculate Accuracy by confusion matrix, Calculate accuracy score, precision score, recall score, f1 score.

### import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### load the data

```python
df = pd.read_csv('Churn.csv')
df.head()
```

### exploratory data analysis

```python
df.info()
```

```python
df.isna().sum()
```

```python
# check if the data is imbalanced
df['Exited'].value_counts()
```

```python
df_minority= df[df['Exited']==1]
df_majority= df[df['Exited']==0]
```

```python
from sklearn.utils import resample

df_up_sampled = resample(df_minority, n_samples=(7963), random_state=123456)
df_up_sampled
```

```python
df_balanced = pd.concat([df_up_sampled, df_majority])
df_balanced
```

```python
df= df_balanced
```

```python
df['Exited'].value_counts()
```

```python
df.drop('Surname',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# encode the geography column
df['Geography'] = LabelEncoder().fit_transform(df['Geography'])

# encode the gender column
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])
```

```python
df.corr()
```

```python
# get all the relationships
correlation = df.corr()['Exited']

# remove the target column as it is 1.0
del correlation['Exited']
```

```python
# get all absolute relationship magnitude
values = []
for value in correlation.values:
    values.append(abs(value))
values
```

```python
# find the min and max values
max = np.max(values)
min = np.min(values)
average_correlation = np.mean([min, max])
print(f"max = {max}, min = {min}, average = {average_correlation}")
```

```python
# find the columns having correlation > average_correlation
columns = []
for index in range(len(correlation.values)):
    if abs(correlation.values[index]) >= average_correlation:
        columns.append(correlation.index[index])
columns
```

### pre-processing

```python
# split the data into x and y

x = df[columns]
y = df['Exited']
x
```

### split the data into train and test sets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=123456)
```

### train the models

#### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

def create_naive_bayes_model():

    # create the model
    model = GaussianNB()

    # train the model
    model.fit(x_train, y_train)

    return model
```

#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

def create_knn_model():

    # create the model
    model = KNeighborsClassifier(n_neighbors=5)

    # train the model
    model.fit(x_train, y_train)

    return model
```

#### SVM

```python
from sklearn.svm import SVC

def create_svm_model():
    # create a model
    model = SVC()
    
    # train the model
    model.fit(x_train, y_train)

    return model
```

### model evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(model_name, model):
    
    # get the predictions using model
    y_pred = model.predict(x_test)
    y_true = y_test

    print(confusion_matrix(y_true, y_pred))
    
    # get the accuracy
    accuracy = accuracy_score(y_true, y_pred)

    # get the precision
    precision = precision_score(y_true, y_pred)

    # get the recall
    recall = recall_score(y_true, y_pred)

    # get f1 score
    f1 = f1_score(y_true, y_pred)

    return model_name, accuracy, precision, recall, f1
```

```python
# create Naive Bayes model
model_nb = create_naive_bayes_model()

# create KNN model
model_knn = create_knn_model()

# create SVM model
model_svm = create_svm_model()
```

```python
# evaluate all the models
model_evaluation_params = pd.DataFrame([
    evaluate_model("Naive Bayes", model_nb),
    evaluate_model('KNN', model_knn), 
    evaluate_model('SVM', model_svm)
], columns=["Model Name", "Accuracy", "Precision", "Recall", "F1 Score"])
```
